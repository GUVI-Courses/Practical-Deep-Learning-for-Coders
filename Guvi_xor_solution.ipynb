```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate XOR data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y = np.array([[0], [1], [1], [0]], dtype=np.float32)

# Function to plot decision boundary
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = np.round(Z).reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), edgecolors="k", cmap=plt.cm.Paired)
    plt.title("Decision Boundary")
    plt.show()

# XOR model without hidden layers (cannot capture non-linearity)
model_without_hidden = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_without_hidden.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_without_hidden = model_without_hidden.fit(X, y, epochs=100, verbose=0)

# XOR model with hidden layers (can capture non-linearity)
model_with_hidden = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_with_hidden.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_with_hidden = model_with_hidden.fit(X, y, epochs=100, verbose=0)

# Plot decision boundaries
print("Decision Boundary without Hidden Layer:")
plot_decision_boundary(model_without_hidden, X, y)

print("Decision Boundary with Hidden Layer:")
plot_decision_boundary(model_with_hidden, X, y)
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generate more XOR data points
X = np.random.uniform(0, 1, (1000, 2))
y = np.logical_xor(X[:, 0] > 0.5, X[:, 1] > 0.5).astype(float).reshape(-1, 1)

# Function to plot decision boundary
def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = np.round(Z).reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
    plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), edgecolors="k", cmap=plt.cm.Paired)
    plt.title("Decision Boundary")
    plt.show()

# XOR model without hidden layers (cannot capture non-linearity)
model_without_hidden = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_without_hidden.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_without_hidden = model_without_hidden.fit(X, y, epochs=100, verbose=0)

# XOR model with hidden layers (can capture non-linearity)
model_with_hidden = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_with_hidden.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history_with_hidden = model_with_hidden.fit(X, y, epochs=100, verbose=0)

# Plot decision boundaries
print("Decision Boundary without Hidden Layer:")
plot_decision_boundary(model_without_hidden, X, y)

print("Decision Boundary with Hidden Layer:")
plot_decision_boundary(model_with_hidden, X, y)
```
