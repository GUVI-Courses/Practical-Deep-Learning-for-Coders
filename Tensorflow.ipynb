# **What is TensorFlow?**

**Open-source library for graph-based numerical computation**

- TensorFlow is a powerful open-source software library developed by the Google Brain Team. It is designed for machine learning and deep learning tasks, making it a popular choice for researchers and practitioners in AI.

**Low and high-level APIs**

- TensorFlow provides a range of APIs for different users:

1. Low-level APIs: Allow operations such as addition, multiplication, and differentiation for mathematical computations.

2. High-level APIs: Simplify the building and training of machine learning models, making it more accessible to beginners.

**Important changes in TensorFlow 2.0**

- TensorFlow 2.0 introduced significant updates to make development faster and more intuitive:

1. Eager execution by default: This enables immediate execution of operations, simplifying debugging and improving code readability.

2. Model building with Keras and Estimators: Keras, a high-level API, is tightly integrated for designing and training deep learning models. Estimators provide a simple, scalable way to develop models for production.

# **What is a Tensor?**

A tensor is an array of data expanding in multiple independent dimensions.

**Generalization of vectors and matrices**

- A tensor is a mathematical object that extends the concept of scalars (single numbers), vectors (1D arrays), and matrices (2D arrays) to higher dimensions. It is a core concept in TensorFlow and is used to represent data.

**Collection of numbers**

- A tensor is essentially a multidimensional array or a collection of numerical values. These numbers can represent anything from raw data to weights in a machine learning model.

**Specific shape**

- Every tensor has a defined shape, which specifies the number of dimensions and the size of each dimension. For example:

1. A scalar is a 0D tensor.
2. A vector is a 1D tensor (e.g., [1, 2, 3]).
3. A matrix is a 2D tensor (e.g., [[1, 2], [3, 4]]).

1. **Scalar**: A single number representing a 0-dimensional tensor.

2. **Vector**: A 1-dimensional tensor with multiple values (e.g., [1, 2]).

3. **Matrix**: A 2-dimensional tensor represented as rows and columns (e.g., [[1, 2], [3, 4]]).

4. **Tensor**: A generalization to higher dimensions, such as a 3D tensor (e.g., [[[1, 2], [3, 2]], [[1, 7], [5, 4]]]).

# **Basic operations**

image.png

In TensorFlow, operations (ops) are the basic building blocks that define computations in a graph. Each operation takes some inputs (e.g., tensors) and produces one or more outputs

**Explanation of the Diagram**

The diagram represents a computational graph in TensorFlow:

**Const Operations:**

Const_1, Const_2, and Const_3 are constant tensors that serve as inputs.
These constants are the starting points of the computation graph.

**Add Operations:**

The Add operation adds two constant tensors (Const_1 and Const_2) to produce a result.
Another Add_1 operation adds the result of Const_2 and Const_3.

**MatMul (Matrix Multiplication):**

The MatMul operation performs a matrix multiplication using the outputs from the Add and Add_1 operations.

````python
import tensorflow as tf

# Define constants
const_1 = tf.constant(3.0, name="Const_1")
const_2 = tf.constant(4.0, name="Const_2")
const_3 = tf.constant(5.0, name="Const_3")

# Add operations
add = tf.add(const_1, const_2, name="Add")
add_1 = tf.add(const_2, const_3, name="Add_1")

# Matrix multiplication operation
# To demonstrate MatMul, tensors must have proper shapes (matrices).
matrix_1 = tf.reshape(add, [1, 1])  # Reshape to 1x1 matrix
matrix_2 = tf.reshape(add_1, [1, 1])  # Reshape to 1x1 matrix
matmul = tf.matmul(matrix_1, matrix_2, name="MatMul")

# Execute the computation
print("Add Result:", add.numpy())
print("Add_1 Result:", add_1.numpy())
print("MatMul Result:", matmul.numpy())
````

**Gradient Computation**

The gradient() operation computes the derivative of a function with respect to its input

````python
import tensorflow as tf

# Define a simple function: f(x) = x^2
def function(x):
    return x**2

# Use GradientTape to compute the gradient
x = tf.Variable(3.0)  # Initialize a variable
with tf.GradientTape() as tape:
    y = function(x)  # Compute f(x)

#tf.GradientTape() is a context manager in TensorFlow used for automatic differentiation
# Compute the gradient dy/dx
gradient = tape.gradient(y, x)
print("Gradient of f(x) = x^2 at x = 3:", gradient.numpy())  # Output: 6.0
````

How It Works:

**Record Operations**: When you execute operations inside the tf.GradientTape() context, TensorFlow automatically "watches" the computations involving trainable variables.

**Compute Gradients**: After exiting the context, you can use the tape.gradient() method to compute the gradients of a target (like a loss function) with respect to these watched variables.

# **A simple dense layer**

````python
import tensorflow as tf

# Define inputs (features)
inputs = tf.constant([[1.0, 35.0]])  # Input vector with two features: [1.0, 35.0]

# Define weights
weights = tf.Variable([[-0.05], [-0.01]])  # Weights for the two features

# Define the bias
bias = tf.Variable([0.5])  # Bias term

# Multiply inputs by the weights (dot product)
product = tf.matmul(inputs, weights)

# Add the bias term
output_with_bias = product + bias

# Apply the sigmoid activation function
dense = tf.keras.activations.sigmoid(output_with_bias)

# Print the results
print("Product (Inputs * Weights):", product.numpy())
print("Output after adding Bias:", output_with_bias.numpy())
print("Dense Layer Output (Sigmoid Activation):", dense.numpy())

````

````python
import tensorflow as tf

# Define the model function
def model(bias, weights, features):
    product = tf.matmul(features, weights)
    return tf.keras.activations.sigmoid(product + bias)

# Compute the predicted values and loss
def loss_function(bias, weights, targets, features):
    predictions = model(bias, weights, features)
    return tf.reduce_mean(tf.keras.losses.binary_crossentropy(targets, predictions))

# Initialize model parameters
bias = tf.Variable([0.0], dtype=tf.float32)  # Bias initialized to 0
weights = tf.Variable(tf.random.normal([20, 1]), dtype=tf.float32)  # Weights for 20 features

# Example data: Borrower features and target values
borrower_features = tf.random.normal([100, 20])  # 100 samples with 20 features
default = tf.random.uniform([100, 1], maxval=2, dtype=tf.int32)  # Binary targets (0 or 1)

# Define the optimizer
opt = tf.keras.optimizers.RMSprop(learning_rate=0.01, momentum=0.9)

# Perform optimization using GradientTape
for step in range(100):  # Run optimization for 100 steps
    with tf.GradientTape() as tape:
        loss = loss_function(bias, weights, default, borrower_features)
    gradients = tape.gradient(loss, [bias, weights])
    opt.apply_gradients(zip(gradients, [bias, weights]))

# Print updated weights and bias after optimization
print("Updated Bias:", bias.numpy())
print("Updated Weights:", weights.numpy())

````

##Building the Model
Now it's time to build the model! We are going to use a keras *sequential* model with three different layers. This model represents a feed-forward neural network (one that passes values from left to right). We'll break down each layer and its architecture below.

````python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),  # input layer (1)
    keras.layers.Dense(128, activation='relu'),  # hidden layer (2)
    keras.layers.Dense(10, activation='softmax') # output layer (3)
])
````

**Layer 1:** This is our input layer and it will conist of 784 neurons. We use the flatten layer with an input shape of (28,28) to denote that our input should come in in that shape. The flatten means that our layer will reshape the shape (28,28) array into a vector of 784 neurons so that each pixel will be associated with one neuron.

**Layer 2:** This is our first and only hidden layer. The *dense* denotes that this layer will be fully connected and each neuron from the previous layer connects to each neuron of this layer. It has 128 neurons and uses the rectify linear unit activation function.

**Layer 3:** This is our output later and is also a dense layer. It has 10 neurons that we will look at to determine our models output. Each neuron represnts the probabillity of a given image being one of the 10 different classes. The activation function *softmax* is used on this layer to calculate a probabillity distribution for each class. This means the value of any neuron in this layer will be between 0 and 1, where 1 represents a high probabillity of the image being that class.

###Compile the Model
The last step in building the model is to define the loss function, optimizer and metrics we would like to track. I won't go into detail about why we chose each of these right now.

````python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

##Training the Model
Now it's finally time to train the model. Since we've already done all the work on our data this step is as easy as calling a single method.

````python
model.fit(train_images, train_labels, epochs=10)  # we pass the data, labels and epochs and watch the magic!
````

##Evaluating the Model
Now it's time to test/evaluate the model. We can do this quite easily using another builtin method from keras.

The *verbose* argument is defined from the keras documentation as:
"verbose: 0 or 1. Verbosity mode. 0 = silent, 1 = progress bar."
(https://keras.io/models/sequential/)

````python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=1)

print('Test accuracy:', test_acc)
````

You'll likely notice that the accuracy here is lower than when training the model. This difference is reffered to as **overfitting**.

And now we have a trained model that's ready to use to predict some values!

##Verifying Predictions
I've written a small function here to help us verify predictions with some simple visuals.

````python
COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR

def predict(model, image, correct_label):
  class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
  prediction = model.predict(np.array([image]))
  predicted_class = class_names[np.argmax(prediction)]

  show_image(image, class_names[correct_label], predicted_class)


def show_image(img, label, guess):
  plt.figure()
  plt.imshow(img, cmap=plt.cm.binary)
  plt.title("Excpected: " + label)
  plt.xlabel("Guess: " + guess)
  plt.colorbar()
  plt.grid(False)
  plt.show()


def get_number():
  while True:
    num = input("Pick a number: ")
    if num.isdigit():
      num = int(num)
      if 0 <= num <= 1000:
        return int(num)
    else:
      print("Try again...")

num = get_number()
image = test_images[num]
label = test_labels[num]
predict(model, image, label)
````
